```python
import pandas as pd
from sklearn.metrics import mean_squared_error
import numpy as np

true = pd.read_csv("C:\\Users\\Afifa\\crypto-challenge\\data\\train.csv")
pred = pd.read_csv("C:\\Users\\Afifa\\crypto-challenge\\submissions\\sample_submission.csv")

# Just for demo (using train subset)
true = true.iloc[-len(pred):]

rmse = np.sqrt(mean_squared_error(true['Target'], pred['prediction']))

print("RMSE:", rmse)
```
